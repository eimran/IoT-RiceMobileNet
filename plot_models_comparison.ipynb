# Models Comparison (Proposed Dataset)

Plots Training Accuracy, Training Loss, and Validation Accuracy for all baseline models, their transfer-learning variants, and RiceMobileNet on the proposed dataset.

```python
import json
import os

import matplotlib.pyplot as plt

BASE_DIR = os.getcwd()
DATASET_DIR = os.path.join(BASE_DIR, "Experiment_on_proposed_DATASET")
BASELINE_DIR = os.path.join(DATASET_DIR, "Baseline_Models", "Model")
BASELINE_TL_DIR = os.path.join(DATASET_DIR, "Baseline_Transfer_learning", "Model")
PROPOSED_HISTORY = os.path.join(
    DATASET_DIR, "Proposed_model", "History", "ProposedModel_history.json"
)
OUTPUT_PATH = os.path.join(
    DATASET_DIR, "Comparison_Plots", "models_comparison_proposed_dataset.png"
)

MODELS = [
    {"label": "CNN", "path": os.path.join(BASELINE_DIR, "CNN", "CNN_history.json"),
     "color": "#1f77b4", "marker": "s", "linestyle": "-", "linewidth": 1.8},
    {"label": "VGG-16", "path": os.path.join(BASELINE_DIR, "VGG16", "VGG16_history.json"),
     "color": "#ff7f0e", "marker": "o", "linestyle": "-", "linewidth": 1.8},
    {"label": "VGG-16_TL", "path": os.path.join(BASELINE_TL_DIR, "VGG16", "VGG16_history.json"),
     "color": "#d62728", "marker": "*", "linestyle": "--", "linewidth": 1.8},
    {"label": "MobileNetV2", "path": os.path.join(BASELINE_DIR, "MobileNetV2", "MobileNetV2_history.json"),
     "color": "#2ca02c", "marker": ">", "linestyle": "-", "linewidth": 1.8},
    {"label": "MobileNetV2_TL", "path": os.path.join(BASELINE_TL_DIR, "MobileNetV2", "MobileNetV2_history.json"),
     "color": "#90ee90", "marker": "<", "linestyle": "--", "linewidth": 1.8},
    {"label": "InceptionV3", "path": os.path.join(BASELINE_DIR, "InceptionV3", "InceptionV3_history.json"),
     "color": "#9467bd", "marker": "D", "linestyle": "-", "linewidth": 1.8},
    {"label": "InceptionV3_TL", "path": os.path.join(BASELINE_TL_DIR, "InceptionV3", "InceptionV3_history.json"),
     "color": "#8c564b", "marker": "D", "linestyle": "--", "linewidth": 1.8},
    {"label": "ResNet-50", "path": os.path.join(BASELINE_DIR, "ResNet50", "ResNet50_history.json"),
     "color": "#e377c2", "marker": "p", "linestyle": "-", "linewidth": 1.8},
    {"label": "ResNet-50_TL", "path": os.path.join(BASELINE_TL_DIR, "ResNet50", "ResNet50_history.json"),
     "color": "#17becf", "marker": "X", "linestyle": "--", "linewidth": 2.2},
    {"label": "RiceMobileNet", "path": PROPOSED_HISTORY,
     "color": "#c9a227", "marker": "*", "linestyle": "-", "linewidth": 2.6},
]

TITLE = "Models Comparison (RLD Dataset)"
MARKER_EVERY = 5

def load_history(path):
    with open(path, "r") as f:
        return json.load(f)


histories = []
for cfg in MODELS:
    if not os.path.exists(cfg["path"]):
        print(f"Warning: missing history file for {cfg['label']}: {cfg['path']}")
        continue
    histories.append((cfg, load_history(cfg["path"])))

fig, (ax_train_acc, ax_train_loss, ax_val_acc) = plt.subplots(1, 3, figsize=(20, 6.5))

shortest_run = min(len(history["accuracy"]) for _, history in histories)

for cfg, history in histories:
    style = dict(
        color=cfg["color"], marker=cfg["marker"], linestyle=cfg["linestyle"],
        linewidth=2.5, markersize=6, markevery=MARKER_EVERY,
    )
    acc, loss, val_acc = history["accuracy"], history["loss"], history["val_accuracy"]
    ax_train_acc.plot(range(1, len(acc) + 1), acc, label=cfg["label"], **style)
    ax_train_loss.plot(range(1, len(loss) + 1), loss, **style)
    ax_val_acc.plot(range(1, len(val_acc) + 1), val_acc, **style)

for ax, ylabel, title in (
    (ax_train_acc, "Train Accuracy (%)", "Training Accuracy"),
    (ax_train_loss, "Train Loss", "Training Loss"),
    (ax_val_acc, "Val Accuracy (%)", "Validation Accuracy"),
):
    ax.axvline(shortest_run, color="violet", linestyle=":", linewidth=2)
    ax.set_xlabel("Epoch", fontsize=12, fontweight="bold")
    ax.set_ylabel(ylabel, fontsize=12, fontweight="bold")
    ax.set_title(title, fontsize=14, fontweight="bold")
    ax.tick_params(axis="both", labelsize=11)
    ax.grid(True, linestyle="--", linewidth=0.7, alpha=0.6)

handles, labels = ax_train_acc.get_legend_handles_labels()
legend = fig.legend(
    handles, labels, loc="upper center", ncol=13, bbox_to_anchor=(0.5, 1.1),
    title=TITLE, fontsize=12, title_fontsize=14, frameon=True, edgecolor="black",
)

legend.get_title().set_fontweight("bold")

fig.tight_layout(rect=[0, 0, 1, 0.95])

os.makedirs(os.path.dirname(OUTPUT_PATH), exist_ok=True)
fig.savefig(OUTPUT_PATH, dpi=300, bbox_inches="tight", format="png")
print(f"Saved figure --> {OUTPUT_PATH}")

plt.show()
```
